```python
import pandas as pd
df=pd.DataFrame
df=pd.read_excel('###Remarks.xlsx')

#df=df.set_index('S No.')

df2=pd.DataFrame(df['Remarks'])
#df2['Remarks'].unique()

df2[df2['Remarks']==0.0]='Invalid Number'
df2[df2['Remarks']==1.0]='Not Interested'
df2[df2['Remarks']==2.0]='May need us in Future'
df2[df2['Remarks']==3.0]='Contact Again'
df2[df2['Remarks']==4.0]='Interested'
#df['Remarks']=df2
#df2
#df2['Remarks'].unique()
#df1=df[['Interest rating ']][df['Remarks']=='Interested']
#df1=df1['Interest rating '].value_counts()
#df1
df['Remarks']=df2
df
```

```python
df2=pd.DataFrame(df['Interest Rating'])
df2[df2['Interest Rating']=='Ready']=4
df2[df2['Interest Rating']=='Good']=2
df2[df2['Interest Rating']=='good']=2
df2[df2['Interest Rating']=='Very Good']=5
df2[df2['Interest Rating']=='Average']=3
df2[df2['Interest Rating']=='god']=2
df2[df2['Interest Rating']=='Eagerly Waiting']=4
df2['Interest Rating'].unique()
df['Interest Rating']=df2
```

```python
df1=df[['Interest Rating']][df['Remarks']=='Interested']
#df1=df1['Interest Rating'].value_counts()
df1['Interest Rating'].unique()
```

```python
df2=df1['Interest Rating'].value_counts()
df2
```

```python
import matplotlib.pyplot as plt
plt.xlabel('Interest Ratings')
plt.ylabel('Frequency')
df2.plot(kind='bar',title='Interest Ratings of the interested companies')
plt.savefig('Interest Ratings graph.png')
```

```python
my_explode = (0, 0, 0,0.2,0)
plt.pie(df2,labels=df2.index,autopct='%1.1f%%',startangle=15, shadow = True,explode=my_explode)
plt.title('Interest Rating Percentage')
plt.axis('equal')
#plt.show()
plt.savefig('Interest Rating Percentage.png')
```

```python
df3=df2['Remarks'].value_counts()
df3
```

```python
import matplotlib.pyplot as plt
plt.xlabel('Remarks')
plt.ylabel('Frequency')
df3.plot(kind='bar',title='Remarks Frequencies')
plt.savefig('Remarks frequencies.png')
```
